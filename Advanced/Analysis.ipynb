```python
import json
with open('tagAn.json', 'r') as f:
    data = json.load(f)
print(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder

dict ={}

for i in data:
  for j in i:
    dict[j] = dict.get(j,0)+1
# print(dict)
dict = {key: value for key, value in dict.items() if key[0]!='*'}



df = pd.DataFrame(dict.items(),columns=['tag','frequency'])
df = df.sort_values(by='frequency', ascending=False)

plt.plot(df['tag'],df['frequency'], marker='o')


plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.show()
```
